```python
# CLEANING DATA TO FIND STATS USED IN LOGISTIC REGRESSION MODEL
# VARIABLES IN MODEL: Fouls home, Fouls away, Fouls in quarter home, Fouls in quarter away, 
# FGM Home, FGM Away, FGA Home, FGA Away, FTM Home, FTM Away, FTA Home, FTA Away, 
# Home Turnovers, Away Turnovers, and Pace
    fouls_h = 0
    fouls_a = 0
    foulsq_h = 0
    foulsq_a = 0
    fgm_h = 0
    fga_h = 0
    fgm_a = 0
    fga_a = 0
    ftm_h = 0
    fta_h = 0
    ftm_a = 0
    fta_a = 0
    turnovers_h = 0
    turnovers_a = 0
    possessions = 0
def comeback_stats(plk_game, pbp):
    comeback_formatz = comeback_format(pbp)
    output = list()
    odds = find_odds(plk_game)
    for i in range(len(comeback_formatz)):
        newrow = list()
        newrow.append(odds[0])
        newrow.append(odds[1])
        newrow.append(round(comeback_formatz[i][0]))
        newrow.append(comeback_formatz[i][1])
        newrow.append(comeback_formatz[i][2])
        newrow.append(comeback_formatz[i][3])
        if comeback_formatz[i][5] != 0:
            newrow.append(round(comeback_formatz[i][4]/comeback_formatz[i][5],2))
        else:
            newrow.append(0)
        newrow.append(comeback_formatz[i][5])
        if comeback_formatz[i][7] != 0:
            newrow.append(round(comeback_formatz[i][6]/comeback_formatz[i][7],2))
        else:
            newrow.append(0)
        newrow.append(comeback_formatz[i][7]) 
        newrow.append(comeback_formatz[i][8])
        newrow.append(comeback_formatz[i][9])
        newrow.append(comeback_formatz[i][10])
        if comeback_formatz[i][12] != 0:
            newrow.append(round(comeback_formatz[i][11]/comeback_formatz[i][12],2))
        else:
            newrow.append(0)
        newrow.append(comeback_formatz[i][12])
        if comeback_formatz[i][14] != 0:
            newrow.append(round(comeback_formatz[i][13]/comeback_formatz[i][14],2))
        else:
            newrow.append(0)
        newrow.append(comeback_formatz[i][14])
        newrow.append(comeback_formatz[i][15])
        newrow.append(comeback_formatz[len(comeback_formatz)-1][16])
        output.append(newrow.copy())
    
    return output


#HELPER FUNCTIONS

def comeback_format(game):
    output = list()
    fouls_h = 0
    fouls_a = 0
    foulsq_h = 0
    foulsq_a = 0
    fgm_h = 0
    fga_h = 0
    fgm_a = 0
    fga_a = 0
    ftm_h = 0
    fta_h = 0
    ftm_a = 0
    fta_a = 0
    turnovers_h = 0
    turnovers_a = 0
    possessions = 0
    
    
    newrow = list()
    for i in range(len(game)):
        newrow = list()
        newrow.append(game[i][0])
        newrow.append(game[i][4])
        if game[i][0] == 600 or game[i][0] == 1200 or game[i][0] == 1800:
            fouls_h = 0
            fouls_a = 0
        if "strata" in game[i][1]:
            possessions += 1
            if action(game[i]) == "home":
                turnovers_h += 1
            else:
                turnovers_a += 1
        if "wolny" in game[i][1]:
            if "wolny 1z2" in game[i][1] or "wolny 1z3" in game[i][1]:
                possessions += 1
            if action(game[i]) == "home":
                fta_h += 1
                if "niecelny" not in game[i][1]:
                    ftm_h += 1
            else:
                fta_a += 1
                if "niecelny" not in game[i][1]:
                    ftm_a += 1
                    
        if "wyskoku" in game[i][1] or "wysoku" in game[i][1] or "lay-up" in game[i][1] or "wsad" in game[i][1]:
            possessions += 1
            if action(game[i]) == "home":
                fga_h += 1
                if "niecelny" not in game[i][1]:
                    fgm_h += 1
            else:
                fga_a += 1
                if "niecelny" not in game[i][1]:
                    fgm_a += 1
        if "faul " in game[i][1]:
            if action(game[i]) == "home":
                fouls_h += 1
                foulsq_h += 1
            else:
                fouls_a += 1
                foulsq_a += 1

        newrow.append(fouls_h)
        newrow.append(foulsq_h)
        newrow.append(fgm_h)
        newrow.append(fga_h)
        newrow.append(ftm_h)
        newrow.append(fta_h)
        newrow.append(turnovers_h)
        newrow.append(fouls_a)
        newrow.append(foulsq_a)
        newrow.append(fgm_a)
        newrow.append(fga_a)
        newrow.append(ftm_a)
        newrow.append(fta_a)
        newrow.append(turnovers_a)
        newrow.append(possessions)
        output.append(newrow.copy())
        
        
    return output


def action(pbp_entry):
        comma_1 = pbp_entry[1].find(",")
        comma_2 = pbp_entry[1][comma_1+1:].find(",")
        player = pbp_entry[1][comma_1+2:comma_2+comma_1+1]
        if player in pbp_entry[6:11]:
            return "home"
        else:
            return "away"


def any_neg_or_zero(temps):
    for i in range(len(temps)):
        if temps[i][3] == 0 or temps[i][3] < 0:
            return True
    return False

def any_pos_or_zero(temps):
    for i in range(len(temps)):
        if temps[i][3] == 0 or temps[i][3] > 0:
            return True
    return False
```

```python
def levenshtein(token1, token2):
    distances = numpy.zeros((len(token1) + 1, len(token2) + 1))

    for t1 in range(len(token1) + 1):
        distances[t1][0] = t1

    for t2 in range(len(token2) + 1):
        distances[0][t2] = t2
        
    a = 0
    b = 0
    c = 0
    
    for t1 in range(1, len(token1) + 1):
        for t2 in range(1, len(token2) + 1):
            if (token1[t1-1] == token2[t2-1]):
                distances[t1][t2] = distances[t1 - 1][t2 - 1]
            else:
                a = distances[t1][t2 - 1]
                b = distances[t1 - 1][t2]
                c = distances[t1 - 1][t2 - 1]
                
                if (a <= b and a <= c):
                    distances[t1][t2] = a + 1
                elif (b <= a and b <= c):
                    distances[t1][t2] = b + 1
                else:
                    distances[t1][t2] = c + 1

    return distances[len(token1)][len(token2)]

def find_odds(plk_game):
    minimum_val = levenshtein(plk21_odds_avg[0][0],plk_game)
    minimum_odds = plk21_odds_avg[0][1:3]
    minimum_matchup = plk21_odds_avg[0][0]
    
    
    for i in range(1,len(plk21_odds_avg)):
        if levenshtein(plk21_odds_avg[i][0],plk_game) < minimum_val:
            minimum_val = levenshtein(plk21_odds_avg[i][0],plk_game)
            minimum_odds = plk21_odds_avg[i][1:3]
            minimum_matchup = plk21_odds_avg[i][0]
    
    #print(minimum_matchup)
    #print(plk_game)
    return minimum_odds
        
```
